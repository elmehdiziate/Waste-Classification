**what is Mask2Former?**
Mask2Former is a universal segmentation model that treats every task (Semantic, Instance, and Panoptic) as a mask classification problem. Instead of predicting a class for every pixel, it predicts a set of binary masks and a class label for each mask.


**The Architecture (How it Works)**
1. The Backbone (Swin-B): * Acts as the "eyes" of the model.

Uses a Swin Transformer (Shifted Windows) to extract features.

It looks at images hierarchically, meaning it can see tiny details (like a bottle cap) and large structures (the whole conveyor belt) simultaneously.

2. The Pixel Decoder: * Takes the raw features from the backbone and refines them.

Uses Multi-scale Deformable Attention to create high-resolution maps.

This ensures the model can accurately trace the complex, thin edges of crushed waste.

3. The Transformer Decoder (The "Secret Agents"): * Uses Object Queries (typically 100). Think of these as "agents" tasked with finding objects.

Masked Attention: Unlike standard models, these agents only look at the area inside their predicted mask. This focus makes the boundaries extremely sharp.

Each query eventually outputs a Binary Mask (the shape) and a Class Score (what is in the shape).

Configs

```python
class_names = [
    'background', 'bottle-blue', 'bottle-blue-full', 'bottle-blue5l',
    'bottle-blue5l-full', 'bottle-dark', 'bottle-dark-full', 'bottle-green',
    'bottle-green-full', 'bottle-milk', 'bottle-milk-full', 'bottle-multicolor',
    'bottle-multicolorv-full', 'bottle-oil', 'bottle-oil-full', 'bottle-transp',
    'bottle-transp-full', 'bottle-yogurt', 'canister', 'cans', 'detergent-box',
    'detergent-color', 'detergent-transparent', 'detergent-white', 'glass-dark',
    'glass-green', 'glass-transp', 'juice-cardboard', 'milk-cardboard'
]

class_colors = [
    [0,0,0], [26,125,226], [4,43,131], [4,36,95], [6,42,185], [0,0,0], [4,10,10],
    [36,179,83], [54,246,37], [72,219,248], [9,65,170], [68,59,112], [104,90,172],
    [221,255,51], [226,253,87], [227,225,235], [243,237,243], [209,255,235],
    [51,51,248], [250,50,83], [39,39,39], [84,84,216], [231,228,231], [248,244,235],
    [21,22,25], [45,250,45], [236,235,250], [126,195,149], [103,199,218]
]

id2label = {i: name for i, name in enumerate(class_names)}
label2id = {name: i for i, name in enumerate(class_names)}
NUM_CLASSES = len(class_names) # 29
```

Class and fucntions

```python
class WarpDataset(Dataset):
    def __init__(self, root_dir, processor):
        self.root = Path(root_dir)
        self.processor = processor
        self.img_dir = self.root / 'JPEGImages'
        self.mask_dir = self.root / 'SegmentationClass'
        self.images = sorted(list(self.img_dir.glob('*.jpg')) + list(self.img_dir.glob('*.png')))

        self.color_map = [
            [0,0,0], [26,125,226], [4,43,131], [4,36,95], [6,42,185], [0,0,0], [4,10,10],
            [36,179,83], [54,246,37], [72,219,248], [9,65,170], [68,59,112], [104,90,172],
            [221,255,51], [226,253,87], [227,225,235], [243,237,243], [209,255,235],
            [51,51,248], [250,50,83], [39,39,39], [84,84,216], [231,228,231], [248,244,235],
            [21,22,25], [45,250,45], [236,235,250], [126,195,149], [103,199,218]
        ]

    def __len__(self):
        return len(self.images)


    #Description: this function maps the RGB colors to class index using self.color_map.
    def rgb_to_id(self, mask_rgb):
        mask_rgb = np.array(mask_rgb)
        h, w, _ = mask_rgb.shape
        mask_id = np.zeros((h, w), dtype=np.uint8)

        for i, color in enumerate(self.color_map):
            match = np.all(mask_rgb == color, axis=-1)
            mask_id[match] = i

        return mask_id

    def __getitem__(self, idx):
        img_path = self.images[idx]
        mask_path = self.mask_dir / (img_path.stem + '.png')

        image = Image.open(img_path).convert("RGB")
        mask_rgb = Image.open(mask_path).convert("RGB")
        mask_np = self.rgb_to_id(mask_rgb)

        inputs = self.processor(images=image, segmentation_maps=mask_np, return_tensors="pt")
        return {
            "pixel_values": inputs["pixel_values"][0],
            "mask_labels": inputs["mask_labels"][0],
            "class_labels": inputs["class_labels"][0]
        }
```

```python
#Description: this method stack only pixel values in one tensor. It keep mask labels and class lables as python lists.
def collate_fn(batch):
    pixel_values = torch.stack([example["pixel_values"] for example in batch])
    mask_labels = [example["mask_labels"] for example in batch]
    class_labels = [example["class_labels"] for example in batch]
    return {
        "pixel_values": pixel_values,
        "mask_labels": mask_labels,
        "class_labels": class_labels
    }
```

```python
#Description: this function maps predicted class Index to RGB values so that we can visualize the prediction as image.
def label_to_color_image(label_map):
    h, w = label_map.shape
    img = np.zeros((h, w, 3), dtype=np.uint8)

    for label_id, color in enumerate(class_colors):
        img[label_map == label_id] = color

    return img
```

```python
import matplotlib.pyplot as plt

#Description: It switched the model from train to eval mode and show a snapshot of model's prediction - then go back to training mode again.
def test_and_show(model, dataset, epoch_num):
    model.eval()
    example = dataset[0]
    pixel_values = example["pixel_values"].unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        outputs = model(pixel_values=pixel_values)

    result = processor.post_process_semantic_segmentation(outputs, target_sizes=[(384, 384)])[0]
    predicted_mask = result.cpu().numpy()
    color_mask = label_to_color_image(predicted_mask)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    img_display = example["pixel_values"].permute(1, 2, 0).cpu().numpy()
    img_display = (img_display - img_display.min()) / (img_display.max() - img_display.min())
    plt.imshow(img_display)
    plt.title(f"Original Image (Epoch {epoch_num})")

    plt.subplot(1, 2, 2)
    plt.imshow(color_mask)
    plt.title("Model Prediction")
    plt.show()

    model.train()
```

```python
#Description: This one allows visualizations of predictions on random number of samples from the dataset.
def run_gallery(dataset, num_samples=5):
    import random
    indices = random.sample(range(len(dataset)), num_samples)

    for idx in indices:
        example = dataset[idx]
        pixel_values = example["pixel_values"].unsqueeze(0).to(DEVICE)

        with torch.no_grad():
            outputs = model(pixel_values=pixel_values)

        target_size = (384, 384)
        result = processor.post_process_semantic_segmentation(outputs, target_sizes=[target_size])[0]
        predicted_mask = result.cpu().numpy()

        fig, ax = plt.subplots(1, 2, figsize=(12, 6))

        img = example["pixel_values"].permute(1, 2, 0).cpu().numpy()
        img = (img - img.min()) / (img.max() - img.min())
        ax[0].imshow(img)
        ax[0].set_title(f"Actual Image (Index {idx})")

        ax[1].imshow(label_to_color_image(predicted_mask))
        ax[1].set_title("Model's Prediction")

        plt.show()
```

Model Training + Visualisation

```python
model_id = "facebook/mask2former-swin-base-ade-semantic"

processor = Mask2FormerImageProcessor.from_pretrained(model_id)
model = Mask2FormerForUniversalSegmentation.from_pretrained(
    model_id,
    num_labels=NUM_CLASSES,
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True
).to(DEVICE)
```

```python
train_dataset = WarpDataset(DATASET_ROOT, processor)
train_loader = DataLoader(
    train_dataset,
    batch_size=2,
    shuffle=True,
    collate_fn=collate_fn
)

optimizer = AdamW(model.parameters(), lr=5e-5, weight_decay=0.05)
model.train()
EPOCHS = 15
for epoch in range(EPOCHS):
    total_loss = 0
    for batch in train_loader:

        pixel_values = batch["pixel_values"].to(DEVICE)
        mask_labels = [l.to(DEVICE) for l in batch["mask_labels"]]
        class_labels = [l.to(DEVICE) for l in batch["class_labels"]]

        optimizer.zero_grad()

        outputs = model(
            pixel_values=pixel_values,
            mask_labels=mask_labels,
            class_labels=class_labels
        )

        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1} | Loss: {total_loss/len(train_loader):.4f}")


    if (epoch + 1) % 5 == 0:
        test_and_show(model, train_dataset, epoch + 1)
```

Metrics

The standard evaluation metric for Semantic Segmentation is mIoU (Mean Intersection over Union)

-- mAP (Mean Average Precision) is an Object Detection metric, hence not used here

```python
import numpy as np
from tqdm import tqdm
from sklearn.metrics import precision_recall_fscore_support

def get_comprehensive_metrics(model, dataset):
    model.eval()
    all_ious = []
    all_precisions = []
    all_recalls = []
    all_f1s = []


    for i in tqdm(range(len(dataset))):
        example = dataset[i]
        pixel_values = example["pixel_values"].unsqueeze(0).to(DEVICE)

        img_path = dataset.images[i]
        mask_path = dataset.mask_dir / (img_path.stem + '.png')
        gt_mask = dataset.rgb_to_id(Image.open(mask_path).convert("RGB"))

        with torch.no_grad():
            outputs = model(pixel_values=pixel_values)

        target_size = gt_mask.shape
        result = processor.post_process_semantic_segmentation(outputs, target_sizes=[target_size])[0]
        pred_mask = result.cpu().numpy()

        gt_flat = gt_mask.flatten()
        pred_flat = pred_mask.flatten()

        p, r, f1, _ = precision_recall_fscore_support(gt_flat, pred_flat, average='weighted', zero_division=0)
        all_precisions.append(p)
        all_recalls.append(r)
        all_f1s.append(f1)

        for cls in range(len(class_colors)):
            intersection = np.logical_and(gt_mask == cls, pred_mask == cls).sum()
            union = np.logical_or(gt_mask == cls, pred_mask == cls).sum()
            if union > 0:
                all_ious.append(intersection / union)

    print('\n')
    print(f"Mean IoU (mIoU):   {np.mean(all_ious):.4f}")
    print(f"Mean Precision:    {np.mean(all_precisions):.4f}")
    print(f"Mean Recall:       {np.mean(all_recalls):.4f}")
    print(f"Mean F1-Score:     {np.mean(all_f1s):.4f}")

get_comprehensive_metrics(model, train_dataset)
```
